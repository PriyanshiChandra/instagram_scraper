```python
def download_img (url,path="C:/Users/Priyanshi Chandra/Desktop/WEB DEV/python_coursera/"):
    resource=urllib.request.urlopen(url)
    filename=path+url[-6:]+'.jpg'
    output=open(filename,"wb")
    output.write(resource.read())
    output.close()
```
